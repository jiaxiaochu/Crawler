```python
# 上海和北京10点到11点的每分钟气温的变化情况
# 导入工具包
from matplotlib import pyplot as plt
import random

# 准备数据
x = range(1,61)
# y = [random.randint(15,18) for i in x]
y_shanghai = [random.randint(15,18) for i in x]
y_beijing = [random.randint(7,14)for i in x]

# 绘制图形
plt.figure(figsize=(20,8),dpi=80)
# plt.plot(x,y)
plt.plot(x,y_shanghai,label="上海",color="#bf3553",linestyle='--')
plt.plot(x,y_beijing,label="北京",color="blue",linestyle='-.')

# 设置x轴的刻度
# x_ticks = [i for i in x]
# plt.xticks(x_ticks[::3])
x_ticks = ["10点{}分".format(i) for i in x]
plt.xticks(list(x)[::3],x_ticks[::3])
# 设置y轴的刻度
x_ticks = []
num = 7
while num <= 18.1:
    x_ticks.append(num)
    num += 0.5
plt.yticks(x_ticks)
# 展示图例
plt.legend(loc=0)

# 设置网格线
plt.grid(alpha=1,linestyle='-.')
# 设置坐标轴信息
plt.title("上海和北京10点到11点的每分钟气温的变化情况")
plt.xlabel("时间")
plt.ylabel("温度")
# 展示图片
plt.show()
```

```python
from matplotlib import pyplot as plt
import random
# 准备数据
x = range(1,61)
y = [random.randint(15,18)for i in x]
y_shanghai = [random.randint(15,18)for i in x]
y_beijing = [random.randint(7,14)for i in x]
# # 设置画布
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(20,8),dpi=80)
# 设置x,y轴信息
x_ticks = ['10点{}分'.format(i)for i in x]
axes[0][0].set_xticks(list(x)[::7])
axes[0][0].set_xticklabels(x_ticks[::7])
axes[0][0].plot(x,y,label='shanghia',color='red',linestyle='-.')
axes[0][1].plot(x,y_shanghai)
axes[1][1].plot(x,y_beijing,label='北京',color='blue',linestyle='-')
#设置坐标描述信息
axes[0][0].legend()
axes[0][0].set_title('上海的温度变化')
axes[1][1].legend()
axes[1][1].grid(linestyle='-.')
plt.show()
```

# 绘制多次柱状图

```python
# 三部电影的首日和首周票房对比
movie_name = ['雷神3：诸神黄昏','正义联盟','寻梦环游记']

first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
# 设置画布
plt.figure(figsize=(10,8),dpi=80)
x = range(len(movie_name))
# 画出柱状图
# 首日票房图
plt.bar(x,first_day,width=0.3,label='首日票房')
# 首周票房图
# 首周柱状图显示的位置在首日的位置右边
plt.bar([i+0.2 for i in x],first_weekend,width=0.3,label='首周票房')
# 显示x轴中文信息，固定首日首周的中间位置
# 设置刻度名称
plt.xticks([i+0.1 for i in x],movie_name)
plt.legend(loc='best')
plt.show()
```

# 绘制饼状图

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.figure(figsize=(20,8),dpi=80)
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
place_count = [60605,54546,45819,28243,13270,9945,7679,6799,6101,4621,20105]
# 通过pie绘图
# autopct小数点前面的数字对产生的结果没有任何影响，小数点后面的数字表示保留小数点几位。
plt.pie(place_count, labels=movie_name, autopct='%1.2f%%', colors=['b','r','g','y','c','m','y','k','c','g','g']);
# 指定显示的pie是正圆
plt.axis('equal')
# plt.legend(loc='upper right')
plt.legend(loc='best')
plt.title('排片占比示意图')
```

# 实现增加阴影、破裂效果饼图

```python
from matplotlib import pyplot as plt
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15,30,45,10]
explode = (0,0.1,0,0)

fig1,ax1 = plt.subplots()
# print(dir(ax1.pie))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal');
```
