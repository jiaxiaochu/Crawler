```python
# 如果我们希望统计电影分类(Actors)的情况，应该如何处理数据？
temp_list = [i.split(",") for i in df["Actors"]]
for j in temp_list:
    print(j)
```

```python
temp_list = [i.split(",") for i in df["Actors"]]
genre_list = np.unique([i.strip() for j in temp_list for i in j])
temp_df = pd.DataFrame(np.zeros((df.shape[0],genre_list.shape[0])),columns=genre_list)
temp_df.shape
```

```python
for i in range(1000):
    temp_list_striped = [j.strip() for j in temp_list[i]]  # Python strip() 方法用于移除字符串头尾指定的字符（默认为空格或换行符）或字符序列。
    print(temp_list_striped)
```

```python
for i in range(1000):
    temp_list_striped = [j.strip() for j in temp_list[i]]  # Python strip() 方法用于移除字符串头尾指定的字符（默认为空格或换行符）或字符序列。
    temp_df.loc[i,temp_list_striped] = 1
print(temp_df.sum().sort_values(ascending=False))
```

```python
temp_df = temp_df.sum().sort_values(ascending=False)[:20]
print(temp_df)
```
