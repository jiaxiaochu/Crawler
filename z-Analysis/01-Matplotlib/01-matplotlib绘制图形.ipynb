```python
# 上海和北京10点到11点的每分钟气温的变化情况
# 导入工具包
from matplotlib import pyplot as plt
import random

# 准备数据
x = range(1,61)
y = [random.randint(15,18) for i in x]
# y_shanghai = [random.randint(15,18) for i in x]
# y_beijing = [random.randint(7,14)for i in x]

# 绘制图形
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)
# plt.show()

# 设置x轴的刻度
# x_ticks = [i for i in x]
# plt.xticks(x_ticks[::3])
x_ticks = ["10点{}分".format(i) for i in x]
plt.xticks(list(x)[::3],x_ticks[::3])
# 设置y轴的刻度
x_ticks = []
num = 7
while num <= 18.1:
    x_ticks.append(num)
    num += 0.5
plt.yticks(x_ticks)
# 展示图例

# 设置网格线
plt.grid(alpha=1,linestyle='-.')
# 设置坐标轴信息
plt.title("上海和北京10点到11点的每分钟气温的变化情况")
plt.xlabel("时间")
plt.ylabel("温度")
# 展示图片
plt.show()
```

```python
# 上海和北京10点到11点的每分钟气温的变化情况
# 导入工具包
from matplotlib import pyplot as plt
import random

# 准备数据
x = range(1,61)
# y = [random.randint(15,18) for i in x]
y_shanghai = [random.randint(15,18) for i in x]
y_beijing = [random.randint(7,14)for i in x]

# 绘制图形
plt.figure(figsize=(20,8),dpi=80)
# plt.plot(x,y)
plt.plot(x,y_shanghai,label="上海",color="#bf3553",linestyle='--')
plt.plot(x,y_beijing,label="北京",color="blue",linestyle='-.')

# 设置x轴的刻度
# x_ticks = [i for i in x]
# plt.xticks(x_ticks[::3])
x_ticks = ["10点{}分".format(i) for i in x]
plt.xticks(list(x)[::3],x_ticks[::3])
# 设置y轴的刻度
x_ticks = []
num = 7
while num <= 18.1:
    x_ticks.append(num)
    num += 0.5
plt.yticks(x_ticks)
# 展示图例
plt.legend(loc=0)

# 设置网格线
plt.grid(alpha=1,linestyle='-.')
# 设置坐标轴信息
plt.title("上海和北京10点到11点的每分钟气温的变化情况")
plt.xlabel("时间")
plt.ylabel("温度")
# 展示图片
plt.show()
```

```python
from matplotlib import pyplot as plt
import random
# 准备数据
x = range(1,61)
y = [random.randint(15,18)for i in x]
y_shanghai = [random.randint(15,18)for i in x]
y_beijing = [random.randint(7,14)for i in x]
# # 设置画布
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(20,8),dpi=80)
# 设置x,y轴信息
x_ticks = ['10点{}分'.format(i)for i in x]
axes[0][0].set_xticks(list(x)[::7])
axes[0][0].set_xticklabels(x_ticks[::7])
axes[0][0].plot(x,y,label='shanghia',color='red',linestyle='-.')
axes[0][1].plot(x,y_shanghai)
axes[1][1].plot(x,y_beijing,label='北京',color='blue',linestyle='-')
#设置坐标描述信息
axes[0][0].legend()
axes[0][0].set_title('上海的温度变化')
axes[1][1].legend()
axes[1][1].grid(linestyle='-.')
plt.show()
```
