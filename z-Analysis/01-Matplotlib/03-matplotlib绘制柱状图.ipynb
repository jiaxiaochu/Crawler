```python
# 完成简单的条形图展现不同的电影票房之间的对比
plt.figure(figsize=(20, 8), dpi=80)

# 准备电影的名字以及电影的票房数据
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
# 放进横坐标的数字列表
x = range(len(movie_name))

# 画出条形图
plt.bar(x, y, width=0.5, color=['b','r','g','y','c','m','y','k','c','g','g'])

# 修改刻度名称
plt.xticks(x, movie_name)

plt.show()
```

## matplolib 绘制多次柱状图

```python
# 三部电影的首日和首周票房对比
movie_name = ['雷神3：诸神黄昏','正义联盟','寻梦环游记']

first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
```

```python
# 设置画布
plt.figure(figsize=(10,8),dpi=80)
x = range(len(movie_name))
# 画出柱状图
# 首日票房图
plt.bar(x,first_day,width=0.3,label='首日票房')
# 首周票房图
# 首周柱状图显示的位置在首日的位置右边
plt.bar([i+0.2 for i in x],first_weekend,width=0.3,label='首周票房')
# 显示x轴中文信息，固定首日首周的中间位置
# 设置刻度名称
plt.xticks([i+0.1 for i in x],movie_name)
plt.legend(loc='best')
plt.show()
```

```python
width = 0.3
plt.figure(figsize=(8,8),dpi=80)

#绘制首日的数据
plt.bar(range(len(movie_name)),first_day,width=width,label="首日")

x_weekend = [i+width for i in range(len(movie_name))]
#绘制首周 的数据
plt.bar(x_weekend,first_weekend,width=width,label="首周")

#设置x轴的刻度
x_ticks = [i+(width/2) for i in range(len(movie_name))]
plt.xticks(x_ticks,movie_name)
plt.legend()
```
