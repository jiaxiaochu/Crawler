## loc 根据索引标签来取值（使用索引的名字）

```python
df
```

```python
df.loc["2018-02-08"]
```

```python
# 一行一列
df.loc["2018-02-08","open"]
```

```python
# 连续多行多列
df.loc["2018-02-27":"2018-02-14","open":"low"]
```

```python
# 不连续的行列
df.loc[["2018-02-27","2018-02-23"],["open","close"]]
```

## iloc根据索引的位置来取值

```python
df
```

```python
# 一行一列
df.iloc[0,0]
```

```python
# 连续的多行多列
df.iloc[:3,:4]
```

```python
# 不连续的行列
df.iloc[[1,3,4],[0,2]]
```
