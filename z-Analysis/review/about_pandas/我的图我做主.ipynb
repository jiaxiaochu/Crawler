数据中共4个字段year（年）、month（月）、gender（性别）、births（出生数）。

使用matplotlib绘制出每年总出生数的柱状图，并使用seaborn设置一种风格。



````python
import pandas as pd
birth=pd.read_excel('../data/births.xlsx')
birth.head()
````

````python
# 根据年份进行分组
group = birth.groupby('year')
group.size()
````

````python
for group_name,group_data in group:
    print(group_name)
````

````python
for group_name,group_data in group:
    print(group_data)
````

````python
from matplotlib import pyplot as plt
import seaborn as sns
# 根据年份进行分组
group = birth.groupby('year')
xticks = []
for group_name,group_data in group:
    xticks.append(group_name)
    plt.bar(group_name,group['births'].sum())
# 设置x轴的显示范围
plt.xticks(xticks)
plt.xlabel('year')
plt.ylabel('births')
sns.set(style='darkgrid')
plt.show()
````
