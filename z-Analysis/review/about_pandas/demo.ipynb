```python
from matplotlib import pyplot as plt

import pandas as pd
df = pd.read_excel('../data/600001SH.xlsx')
df.index  = pd.to_datetime(df['日期'])
feb_datas = df['2015-02']

fig = plt.figure(figsize=(20,8),dpi=80)
ax = fig.add_subplot(111)

lin1 = ax.plot(feb_datas.index.tolist(),feb_datas['开盘价(元)'].tolist(),label='开盘价',color='red',linestyle="-.")
ax.set_ylabel('开盘价(元)')
ax.set_xlabel('日期')
# 展示图例
ax.legend(loc=0)  

ax2 = ax.twinx()
line2 = ax2.plot(feb_datas.index.tolist(),feb_datas['成交量(股)'],label='成交量(股)')
ax2.set_ylabel('成交量(股)')
ax2.legend(loc=0)
# plt.plot(feb_datas.index.tolist(),feb_datas['开盘价(元)'].tolist())

lns = lin1 + line2
labs = [l.get_label() for l in lns]
ax.legend(lns,labs)

ax.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

y = range(1,14,2)
x = [-3,-2,-1,0,1,2,3]

# 获取当前图表的图像
ax = plt.gca()

# 设置图形的包围线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

y = range(1,14,2)
x = [-3,-2,-1,0,1,2,3]

# 获取当前图表的图像
ax = plt.gca()

# 设置图形的包围线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('yellow')

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

y = range(1,14,2)
x = [-3,-2,-1,0,1,2,3]

# 获取当前图表的图像
ax = plt.gca()

# 设置图形的包围线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('yellow')

ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.plot(x,y)
plt.show()
```
