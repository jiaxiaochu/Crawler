# 设置中文显示

```python
# `from matplotlib import font_manager`导入字体管理模块，
# `font_manager.FontProperties()`使用这个方法根据我提供的字体文件创建字体对象。
# 当需要有中文显示的地方，添加字体属性就可以了。

from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='../data/STSONG.TTF')
```

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
import random
# 创建字体对象
my_font = font_manager.FontProperties(fname='../data/STSONG.TTF',size=18)
x = range(0,120)
y = [random.randint(10,30) for i in range(120)]
# 添加字体属性  
# 增加标题、x轴y轴描述信息  设置字体
plt.ylabel("次数",fontproperties=my_font)
plt.xlabel("时间",fontproperties=my_font)
# 设置标题
plt.title('每分钟跳动次数',fontproperties=my_font)
plt.plot(x,y)
plt.show()
```

# 自定义X轴刻度

```python
import matplotlib.pyplot as plt
import random
# 绘制基础折线图
x = range(1,61)
y = [random.randint(15,18)for i in x]
# 绘制图形
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)
# plt.show()

# 设置x轴刻度
x_ticks = ["10点{}分".format(i)for i in x] # 刻度的值为中文
plt.xticks(list(x)[::3],x_ticks[::3],rotation=45) # 设置x的轴的刻度和稀疏程度
plt.show()
```

```python
xticks(locs, [labels], **kwargs) 
locs参数为数组参数，表示x-axis的刻度线显示标注的地方，即ticks放置的地方，
第二个参数也为数组参数，表示在locs数组表示的位置添加的标签。
# list(x)[::3]为xticks()的第一个参数，根据x值的个数调整x轴的刻度疏密程度。
# ],x_ticks[::3],为xticks()的第二个参数，还是使用x的值作为刻度的标签值，但是这里获取了其中的一部分，确保第一个参数和第二个参数的个数相同。
# `rotation=45`默认刻度的值是横向书写的，这样会有一定的重叠，所以我们将文字进行旋转操作，45位旋转的度数。
```

```python
import matplotlib.pyplot as plt
import random
random.seed(10)  # 设置随机种子，保证每次随机的结果相同
plt.figure(figsize=(20,8),dpi=80)
x = range(60)
y = [random.randint(15,20)for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# plt.xticks(x[::5]) #设置x的轴的刻度和稀疏程度
plt.xticks(x[::5],x_ticks[::5]) # 设置x的轴的刻度值为中文和熟悉程度
plt.grid(alpha=0.5,linestyle="-.") # 设置网格和网络的粗细和网络线条的样式
plt.plot(x,y)
```

# 一图多个坐标系子图

```python
from matplotlib import pyplot as plt
import random
```

```python
# 设置画布
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(20,8),dpi=80)
plt.show()
# plt.subplot(nrows, ncols, index)
# 函数的 `nrows` 参数指定将数据图区域分成多少行；`ncols` 参数指定将数据图区域分成多少列；`index` 参数指定获取第几个区域。
```

```python
from matplotlib import pyplot as plt
import numpy as np
x = [1,10,14,15,16,17]
y = np.array([3,4,6,2,1,5])

plt.figure(figsize = (10,8))
# 第一个子图
# 折线图
plt.subplot(2, 2, 1)
plt.plot(y)
plt.title('Axes1')
#第二个子图
# 折线图，y轴每个数据的立方
plt.subplot(2, 2, 2)
plt.plot(y**3)
plt.title('Axes2')
#第三个子图
# 折线图，x轴和y轴均指定数据
plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title('Axes3')
plt.show()

# 我们创建一个包含2行2列，共4个子图。
# 第一个子图的创建方法是plt.subplot(2, 2, 1)，
# 第二个子图就使用：plt.subplot(2, 2, 2)，以此类推。
```

```python
from matplotlib import pyplot as plt
import random
# 准备数据
x = range(1,61)
y = [random.randint(15,18)for i in x]
y_shanghai = [random.randint(15,18)for i in x]
y_beijing = [random.randint(7,14)for i in x]
# # 设置画布
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(20,8),dpi=80)
# 设置x,y轴信息
x_ticks = ['10点{}分'.format(i)for i in x]
axes[0][0].set_xticks(list(x)[::7])
axes[0][0].set_xticklabels(x_ticks[::7])
axes[0][0].plot(x,y,label='shanghia',color='red',linestyle='-.')
axes[0][1].plot(x,y_shanghai)
axes[1][1].plot(x,y_beijing,label='北京',color='blue',linestyle='-')
#设置坐标描述信息
axes[0][0].legend()
axes[0][0].set_title('上海的温度变化')
axes[1][1].legend()
axes[1][1].grid(linestyle='-.')
plt.show()
```
