## 累计统计函数

```python
df = df.sort_index()
```

```python
df
```

```python
df["p_change"].cumsum()
```

```python
# cumsum	计算前1/2/3/…/n个数的和
# cummax	计算前1/2/3/…/n个数的最大值
# cummin	计算前1/2/3/…/n个数的最小值
# cumprod	计算前1/2/3/…/n个数的积
df["p_change"].cumsum().plot(figsize=(20,8),fontsize=20)
```

## pandas的逻辑运算

```python
df
```

```python
# df[df["p_change"]>2]
df["p_change"]>2
```

```python
df[(df["p_change"]>1.5) & (df["p_change"]<2)]
```

```python
# isin 判断值是否在列表中，如果在返回True，否则False

df[df["p_change"].isin([2.62,1.44])]  #选择p_change的值为2.62或者是1.44的所有的行
```

```python
df["p_change"].add(1)
```

```python
high = df["high"]
low = df["low"]
high.sub(low).sort_values()
```

```python
# apply进行自定义的运算
df[["high","low"]].apply(lambda x:x.max()-x.min(),axis=1)
```
