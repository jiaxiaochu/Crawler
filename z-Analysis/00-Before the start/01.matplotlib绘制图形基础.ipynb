## 折线图

```python
from matplotlib import pyplot as plt
import random
```

```python
# 准备数据
x = range(1,31)
y = [random.randint(10,20)for i in x]
# 开始绘制图形
# 设置图形的大小
plt.figure(figsize=(20,8),dpi=80)
## plot能够绘制折线图
plt.plot(x,y)
## 保存图片
plt.savefig("./气温.png")
plt.show()
```

## 设置xy轴的刻度

```python
# 1.0
# 准备数据
x = range(1,61)
y = [random.uniform(15,18) for i in x]

# 绘制图形
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)

#设置x轴的刻度
x_ticks = ["10点{}分".format(i) for i in x]
plt.xticks(range(1,40,3))

# 设置Y轴的刻度
x_ticks = []
num = 15
while num <= 18.1:
    x_ticks.append(num)
    num+=0.3
plt.yticks(range(1,40))
plt.show()
```

```python
# 2.0
# 准备数据
x = range(1,61)
y = [random.uniform(15,18) for i in x]

# 绘制图形
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)

#设置x轴的刻度
x_ticks = ["10点{}分".format(i) for i in x]
plt.xticks(list(x)[::3],x_ticks[::3])
# plt.xticks(range(1,40,3))

# 设置Y轴的刻度
x_ticks = []
num = 15
while num <= 18.1:
    x_ticks.append(num)
    num+=0.3
plt.yticks(x_ticks)
#plt.yticks(range(1,40))
plt.show()
```
