## 自定义x,y刻度以及中文显示

```python
from matplotlib import pyplot as plt
import random
```

```python
random.seed(10)  # 设置随机种子，保证每次随机的结果相同
plt.figure(figsize=(20,8),dpi=80)
x = range(60)
y = [random.randint(15,20)for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5]) #设置x的轴的刻度和稀疏程度
plt.grid(alpha=0.5,linestyle="-.") # 设置网格和网络的粗细和网络线条的样式
plt.plot(x,y)
```

## 设置中文

```python
random.seed(10)  # 设置随机种子，保证每次随机的结果相同
plt.figure(figsize=(20,8),dpi=80)
x = range(60)
y = [random.randint(15,20)for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# plt.xticks(x[::5]) #设置x的轴的刻度和稀疏程度
plt.xticks(x[::5],x_ticks[::5]) # 设置x的轴的刻度值为中文和熟悉程度
plt.grid(alpha=0.5,linestyle="-.") # 设置网格和网络的粗细和网络线条的样式
plt.plot(x,y)
```

## 设置图形的展示信息

```python
from matplotlib import pyplot as plt
import random
```

```python
# 准备数据
x = range(1,61)
y_shanghai = [random.randint(15,18)for i in x]
y_beijing = [random.randint(7,14)for i in x]
# 绘制图形
plt.figure(figsize=(20,8), dpi=80)
plt.plot(x,y_shanghai,label="上海1",color='#3CB371',linestyle='——')
plt.plot(x,y_beijing,label="北京",color="#FF69B4",linestyle="—.")

# 设置X轴的刻度
x_ticks = ["10点{}分".format(i) for i in x]
plt.xticks(list(x)[::3], x_ticks[::3])

#设置y轴的刻度
x_ticks = []
num = 7
while num <= 18.1:
    x_ticks.append(num)
    num+=0.5
plt.yticks(x_ticks)


#展示图例
plt.legend(loc=0)

#设置显示网格
plt.xlabel("时间")
plt.ylable("温度")

plt.title("上海和北京10点到11点的每分钟气温的变化情况")
plt.show()
```

```python
#准备数据
x = range(1,61)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(7,14) for i in x]

#绘制图形
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_shanghai,label="上海1",color="#3CB371",linestyle="--")
plt.plot(x,y_beijing,label="北京",color="#FF69B4",linestyle="-.")

#设置x轴的刻度
x_ticks = ["10点{}分".format(i) for i in x]
plt.xticks(list(x)[::3],x_ticks[::3])
# plt.xticks(range(1,40,3))
#设置y轴的刻度
x_ticks = []
num = 7
while num<=18.1:
    x_ticks.append(num)
    num+=0.5
plt.yticks(x_ticks)
#展示图例
plt.legend(loc=0)

#设置显示网格
plt.grid(alpha=0.3,linestyle="-.")

plt.xlabel("时间")

plt.ylabel("温度")
plt.title("上海和北京10点到11点的每分钟气温的变化情况")
plt.show()
```
