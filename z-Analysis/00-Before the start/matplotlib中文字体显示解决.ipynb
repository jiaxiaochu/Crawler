## 解决matplotlib中文字体显示问题

```python
import matplotlib.pyplot as plt
import random
random.seed(10)  # 设置随机种子，保证每次随机的结果相同
plt.figure(figsize=(20,8),dpi=80)
x = range(60)
y = [random.randint(15,20)for i in x]
x_ticks = ["11点{}分".format(i) for i in x]
# plt.xticks(x[::5]) #设置x的轴的刻度和稀疏程度
plt.xticks(x[::5],x_ticks[::5]) # 设置x的轴的刻度值为中文和熟悉程度
plt.grid(alpha=0.5,linestyle="-.") # 设置网格和网络的粗细和网络线条的样式
plt.plot(x,y)
```
