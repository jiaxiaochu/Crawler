```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
path = "/Users/jiazhixiang/doc/python-data-analysis/data/US_video_data_numbers.csv"
```

```python
#["views","likes","dislikes","comment_total"]
t1 = np.genfromtxt(path,delimiter=',')
```

```python
t1
```

```python
# 获取评论数据
comment_array = t1[:,-1]   # 获取最后一列的值
comment_array = comment_array[comment_array<40000]
comment_array
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.hist(comment_array,bins=20);
```

```python
print(range(comment_array.shape[0]))
```

```python
plt.figure(figsize=(20,8),dpi=80);
plt.scatter(range(comment_array.shape[0]),comment_array);
plt.xticks(fontsize=20);
plt.yticks(fontsize=20);
```

```python
# 视屏的评论数和喜欢数的关系
x_data = t1[:,-1]  # 评论数
y_data = t1[:,1]  # 喜欢数
# 删除评论数中的离群数据
x = x_data[x_data < 50000]
print(x)
y = y_data[x_data < 50000]
print(y)
plt.figure(figsize=(20,8),dpi=80);
plt.xticks(fontsize=20);
plt.yticks(fontsize=20);
plt.scatter(x,y);
```
