```python
# 使用numpy生成股票数据
t1 = np.random.standard_normal((500,504))
t1
```

```python
# pd.date_range()方法的介绍
pd.date_range(start = "20180101",end="20180606",freq="B")  # 指定结束日期
```

```python
# 给DataFrame添加行索引
index = ["stock_{}".format(i) for i in range(t1.shape[0])]
columns = pd.date_range(start = "20170101",periods = t1.shape[1], freq="B")
columns = [i.strftime("%Y-%m-%d")for i in columns]
print(columns[:10])
t2 = pd.DataFrame(t1,index=index,columns=columns)
```
